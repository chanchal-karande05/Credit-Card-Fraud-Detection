```python
import seaborn as sns

plt.figure(figsize=(15, 10))

# 1. Amount
plt.subplot(2,2,1)
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title('Transaction Amount')

# 2. Hourly (CRITICAL for fraud patterns)
df['Hour'] = (df['Time']/3600)%24
plt.subplot(2,2,2)
sns.histplot(df['Hour'], bins=24)
plt.title('Hourly Pattern')

# 3. Class imbalance
plt.subplot(2,2,3)
sns.countplot(x='Class', data=df)
plt.title(' Fraud Normal')

# 4. Fraud amount
plt.subplot(2,2,4)
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Fraud Amounts Higher')
```

DATA PREPROCESSING

```python
features = df.drop(['Class','Hour'],axis=1)

X = features.copy()
```

```python
X.shape
```

```python
df.shape
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X['Scaled_Time'] = scaler.fit_transform(X[['Time']])
X['Scaled_Amount'] = scaler.fit_transform(X[['Amount']])

X = X.drop(['Time', 'Amount'], axis=1)
print("X Shape:",X.shape)
```

```python
X
```

ISOLATION FOREST TRAINING

```python
from sklearn.ensemble import IsolationForest

model = IsolationForest(n_estimators=1000,contamination=0.0017,random_state=42,n_jobs=-1)
model.fit(X)

predictions = model.predict(X)
fraud_count = sum(predictions == -1)
print(f"Number of Fradulent Transactions : {fraud_count}")
print(f"Fradulent percentage :{fraud_count/len(X)*100:.3f}%")
```

MODEL EVALUATION

```python
import numpy as np
```

```python
Y_pred = np.where(predictions == -1,1,0)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print("Classification Report:")
print(classification_report(df['Class'],Y_pred,target_names=["Normal","Fraud"]))

print("Confusion Matrix:")
print(confusion_matrix(df['Class'],Y_pred))
```

```python
import joblib

final_model = IsolationForest(
   n_estimators=1000,
   contamination=0.0017,
   random_state=42,
   n_jobs=-1
)

final_model.fit(X)
joblib.dump(final_model,'fraud_detection.pkl')

print("Production Model Saved")
```
